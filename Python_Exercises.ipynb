```python
# Initialisation de la liste avec 5 entiers
liste_nombres = [4, 7, 15, 23, 42]

# Saisie d'un entier avec input()
while True:
    try:
        nombre_saisi = int(input("Entrez un entier : "))
        break
    except ValueError:
        print("Veuillez entrer un nombre entier valide.")

# Vérification si l'entier est dans la liste
if nombre_saisi in liste_nombres:
    print(f"Le nombre {nombre_saisi} a été trouvé dans la liste !")
else:
    print(f"Le nombre {nombre_saisi} n'est pas dans la liste.")

# Saisie d'un autre entier positif
while True:
    try:
        nombre_test = int(input("Entrez un entier positif : "))
        if nombre_test > 1:
            break
        else:
            print("Veuillez entrer un nombre supérieur à 1.")
    except ValueError:
        print("Veuillez entrer un nombre entier valide.")

diviseur = None
i = 2

# Vérification si le nombre est premier
while i * i <= nombre_test:  # Test jusqu'à la racine carrée de nombre_test
    if nombre_test % i == 0:
        diviseur = i
        print(f"Le nombre {nombre_test} n'est pas premier. Son premier diviseur est {diviseur}.")
        break
    i += 1
else:
    print(f"Le nombre {nombre_test} est premier !")
```

```python
import random

while True:
    def RockPaperScissors():
      your_choice = input("Enter your choice (rock as R, paper as P, scissors as S): ").lower()
      computer_choice = random.choice(["R", "P", "S"])

      if your_choice == computer_choice:
        print("It's a tie!")

      elif (your_choice == "R" and computer_choice == "S"):
        print("You win!")

      elif (your_choice == "P" and computer_choice == "R"):
        print("You win!")

      elif (your_choice == "S" and computer_choice == "P"):
        print("You win!")

      else:
        print("You lose!")
        print(f"Computer choice: {computer_choice}")

      exit = input("Do you want to exit ?").lower()
    if exit == "yes":
        break

    RockPaperScissors()
```
