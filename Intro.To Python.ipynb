```python
import random

def Rock_Paper_Scissor_Game():
    while True:
        # Take user input and convert it to lowercase for case-insensitive comparison
        MainChoice = input("Rock, Paper, or Scissor? ").strip().lower()

        # Validate input
        if MainChoice not in ['rock', 'paper', 'scissor']:
            print('Invalid Choice')
            break

        # Random choice by the computer
        Random = random.choice(['rock', 'paper', 'scissor'])

        # Display choices
        print(f'You chose: {MainChoice.capitalize()}')
        print(f'Computer chose: {Random.capitalize()}')

        # Determine the winner
        if MainChoice == Random:
            print('Draw')
        elif (MainChoice == 'rock' and Random == 'scissor') or \
             (MainChoice == 'paper' and Random == 'rock') or \
             (MainChoice == 'scissor' and Random == 'paper'):
            print('You Win')
        else:
            print('You Lose')

        # Ask if the user wants to play again
        play_again = input("Do you want to play again? (yes/no): ").strip().lower()
        if play_again != 'yes':
            break

Rock_Paper_Scissor_Game()
```

```python
MyStack = []
stacksize = 3

def DisplayStack():
    print("État courant de la pile :")
    for Item in MyStack:
        print(Item)

def Push(Value):
    if len(MyStack) < stacksize:
        MyStack.append(Value)
        print(f"{Value} ajouté")
    else:
        print("La pile est pleine !")

def Pop():
    if len(MyStack) > 0:
        Removed = MyStack.pop()
        print(f"{Removed} supprimé de la pile")
    else:
        print("La pile est vide.")

Push(1)
Push(2)
Push(3)
DisplayStack()
input("Appuyez sur une touche quand vous êtes prêt...")
Push(4)
DisplayStack()
input("Appuyez sur une touche quand vous êtes prêt...")
Pop()
DisplayStack()
input("Appuyez sur une touche quand vous êtes prêt...")
Pop()
Pop()
DisplayStack()
```

```python
import collections

# Step 1: Create a deque with initial elements and a maximum length
MyDeque = collections.deque("abcdef", 10)

# Step 2: Print the initial state of the deque
print("État initial :")
for Item in MyDeque:
    print(Item, end=" ")
print("\r\n")

# Step 3: Add elements to the right
print("Ajout et extension à droite")
MyDeque.append("h")
MyDeque.extend("ij")
for Item in MyDeque:
    print(Item, end=" ")
print("\r\nMyDeque contient {0} éléments.".format(len(MyDeque)))

# Step 4: Remove elements from the right
print("\r\nPop à droite")
print("Pop de {0}".format(MyDeque.pop()))
for Item in MyDeque:
    print(Item, end=" ")
print("\r\n")

# Step 5: Add elements to the left
print("Ajout et extension à gauche")
MyDeque.appendleft("a")
MyDeque.extendleft("bc")
for Item in MyDeque:
    print(Item, end=" ")
print("\r\nMyDeque contient {0} éléments.".format(len(MyDeque)))

# Step 6: Remove elements from the left
print("\r\nPop à gauche")
print("Pop de {0}".format(MyDeque.popleft()))
for Item in MyDeque:
    print(Item, end=" ")
print("\r\n")

# Step 7: Remove a specific element
print("Suppression")
MyDeque.remove("a")
for Item in MyDeque:
    print(Item, end=" ")
print("\r\n")
```

```python
class MyClass:
  def PrintList1(*args):
    for Count, Item in enumerate(args):
      print("{0}. {1}".format(Count, Item))

  def PrintList2(**kwargs):
    for Name, Value in kwargs.items():
      print("{0} likes {1}".format(Name, Value))

MyClass.PrintList1("Red", "Blue", "Yellow", "Green")
MyClass.PrintList2(Annie="Red", Steve="Blue", Sarah="Yellow")
```

```python
class MyClass:
  def __init__(self, *args):
    self.Input = args

  def __add__(self, Other):
    Output = MyClass()
    Output.Input = self.Input + Other.Input
    return Output

  def __str__(self):
    Output = ""
    for Item in self.Input:
      Output += Item
      Output += " "
    return Output

List1 = MyClass("Red", "Green", "Blue")
List2 = MyClass("Purple", "Yellow", "Orange")
List3 = List1 + List2
print("{0} + {1} = {2}".format(List1, List2, List3))
```
