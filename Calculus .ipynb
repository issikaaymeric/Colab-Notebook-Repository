```python
sp.integrate(sp.csc(x) * sp.cot(x), x)
```

```python
integral = sp.integrate(8*x + sp.csc(x)**2, x)
```

```python
C = - integral.subs(x, sp.pi/2) -17
y = integral + C
```

```python
y.subs(x, sp.pi/2)
```

```python
sp.integrate((1 + sp.sqrt(x)**sp.Rational(1, 3))/ sp.sqrt(x), x)
```

```python
sp.integrate(x*(1 - x**2)**sp.Rational(1, 4), x)
```

```python
sp.integrate((2*x - 1)*sp.cos(sp.sqrt(3*(2*x - 1)**2 + 6)) / sp.sqrt(3*(2*x - 1)**2 + 6), x)
```

```python
sp.integrate(sp.exp(x) / sp.sqrt(sp.exp(2*x) + 9), (x, 0, sp.log(4)))
```
