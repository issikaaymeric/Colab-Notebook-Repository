```python
import pandas as pd

data = [10, 20, 30, 40, 50]
items = ["coffee", "shoes", "clothes", "phone", "laptop"]

#s = pd.Series(data).astype("float")
s = pd.Series(data, index=items, name="Sales")
print(s)

# Corrected: Use square brackets and a single index for a Series
position = s.iloc[4]
print("____"*10)
print(position)

print("____"*10)

item_position = s.loc["laptop"]
print(item_position)

item_index = s.index[3], int(s.iloc[3])
print("____"*10)
print(item_index)

print("____"*10)

s.mean()
s.median()
s.std()
```
