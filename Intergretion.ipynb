```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = x**2 + 2*x - 6
```

```python
f_lambdified = sp.lambdify(x, f, 'numpy')
x_values = np.linspace(-1, 10, 100)
y_values = f_lambdified(x_values)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='f(x)', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
f_prime = sp.diff(f, x)
f_prime_lambdified = sp.lambdify(x, f_prime, 'numpy')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_values, f_prime_lambdified(x_values), label = 'f\'(x)', color = 'blue')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('Graph of f\'(x)')
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = x**3 * sp.exp(x) * sp.sin(10*x)**5

f_lambdified = sp.lambdify(x, f, "numpy")
values_x = np.linspace(-1, 10, 400)
values_y = f_lambdified(values_x)

plt.figure(figsize=(15, 10))
plt.plot(values_x, values_y, label="f(x)", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x)")
plt.grid(True)
plt.legend()

f_prime = sp.diff(f, x)
f_prime_lambdified = sp.lambdify(x, f_prime, "numpy")

plt.figure(figsize=(15, 10))
plt.plot(values_x, f_prime_lambdified(values_x), label="f'(x)", color="red")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Graph of f'(x)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
from scipy.integrate import quad

def f(x):
  return 12 + 3*x

def g(x):
  return 2*x**2 + 3

firt_result = quad(f, 1, 3)
second_result = quad(g, 1, 3)
result = np.array(firt_result) - np.array(second_result)
print(f"Your result is : {result[0]:.2f}")
```
