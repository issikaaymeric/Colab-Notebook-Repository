```python
from statsbombpy import sb

# Fetch available competitions and their seasons
competitions = sb.competitions()

# Print the competitions to identify valid IDs
display(competitions)

# Select a valid competition_id and season_id
# For example, you can find the competition and season IDs for the
# "FA Women's Super League" in the competitions DataFrame
competition_id = 37
season_id = 42

# Now fetch the matches with the valid IDs
matches = sb.matches(competition_id=competition_id, season_id=season_id)

# Display the matches DataFrame
display(matches)
```
