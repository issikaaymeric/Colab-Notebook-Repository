```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = x**2 - 2*x + 1
```

```python
f_lambdified = sp.lambdify(x,f, 'numpy')
x_vals = np.linspace(-10, 10, 400)
y_vals = f_lambdified(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.title("Graph of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)
critical_points = sp.solve(f_prime, x)
second_derivative_values = [f_double_prime.subs(x, pt) for pt in critical_points]

print("\n--- Derivatives ---")
print("f'(x):", f_prime)
print("f''(x):", f_double_prime)

print("\n--- Critical Points ---")
for pt, second_val in zip(critical_points, second_derivative_values):
    nature = "Min" if second_val > 0 else "Max" if second_val < 0 else "Inflection"
    print(f"x = {pt}, f(x) = {f.subs(x, pt)}, Type: {nature}")

# STEP 4: Intercepts
y_intercept = f.subs(x, 0)
x_intercepts = sp.solve(f, x)
print("\n--- Intercepts ---")
print("Y-intercept:", y_intercept)
print("X-intercepts:", x_intercepts)

# STEP 5: Limits
limit_neg_inf = sp.limit(f, x, -sp.oo)
limit_pos_inf = sp.limit(f, x, sp.oo)
print("\n--- Limits ---")
print("Limit as x -> -∞:", limit_neg_inf)
print("Limit as x -> ∞:", limit_pos_inf)
```
