````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.colors import ListedColormap, BoundaryNorm

# Define the mapping
value_map = {'P': 1, 'G': 2, 'B': 3, 'Bst': 4}

# Create the table
row_labels = [' P', ' G', ' B', ' Bst']
col_labels = [' P', ' G', ' B', ' Bst']

# Create the matrix based on multiplication rule
data = []
for row in ['P', 'G', 'B', 'Bst']:
    row_data = []
    for col in ['P', 'G', 'B', 'Bst']:
        result = value_map[row] * value_map[col]
        row_data.append(result)
    data.append(row_data)

# Create DataFrame
df = pd.DataFrame(data, index=row_labels, columns=col_labels)

# Define colors for risk levels
# Scores range from 1 (lowest) to 16 (highest)
# Green (low risk) -> Light Green -> Yellow (medium) -> Red (high risk)
colors = ['#ff0000', '#ffff00', '#90ee90', '#00cc00']
boundaries = [0.5, 2.5, 6.5, 10.5, 16.5]  # Boundaries for scores 1-2, 3-6, 7-10, 11-16
cmap = ListedColormap(colors)
norm = BoundaryNorm(boundaries, cmap.N)

# Create the heatmap
plt.figure(figsize=(10, 8))
ax = sns.heatmap(df, annot=True, fmt='d', cmap=cmap, norm=norm,
                 cbar_kws={'label': 'Risk Score'},
                 linewidths=2, linecolor='white',
                 square=True, vmin=1, vmax=16)

# Customize colorbar
cbar = ax.collections[0].colorbar
cbar.set_ticks([1.5, 4.5, 8.5, 13.5])
cbar.set_ticklabels(['Low (1-2)', 'Low-Med (3-6)', 'Medium (7-10)', 'High (11-16)'])

plt.title('Project Management Risk Assessment Matrix\n(Multiplication Rule: P=1, G=2, B=3, Bst=4)',
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Probability', fontsize=12, fontweight='bold')
plt.ylabel('Impact', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

# Print the risk matrix for reference
print("\nRisk Score Matrix:")
print(df)
print("\nRisk Level Legend:")
print("1-2: Low Risk (Green)")
print("3-6: Low-Medium Risk (Light Green)")
print("7-10: Medium Risk (Yellow)")
print("11-16: High Risk (Red)")
````

**TCP (Transmission Control Protocol):**
- Connection-oriented protocol (establishes connection before data transfer)
- Provides reliable, ordered delivery with error checking and acknowledgments
- Example: Web browsing (HTTP/HTTPS), Email (SMTP), File transfer (FTP)

**UDP (User Datagram Protocol):**
- Connectionless protocol (no connection establishment)
- No guaranteed delivery, faster with less overhead
- Example: Video streaming, VoIP, DNS queries, online gaming

**When to choose:**
- TCP: When reliability and data integrity are critical (file transfers, web pages, emails)
- UDP: When speed is more important than perfect reliability (live video, gaming, DNS)

**Model Answer:**

1. **Subnet Mask (2 points):** A 32-bit number that defines which portion of an IP address represents the network portion and which represents the host portion. It's used by devices to determine if a destination IP address is on the local network (can communicate directly) or on a remote network (must send to default gateway). **b) Subnet scheme (4 points):**

**Given:** 192.168.0.0/23 = 192.168.0.0 to 192.168.1.255 (512 total addresses)

- Network Address/CIDR: **192.168.0.0/26**

#### 2. OSI and TCP/IP Models

**Memorization Mnemonics:**

  - Physical, Data Link, Network, Transport, Session, Presentation, Application
- **OSI Layers (top to bottom):** "All People Seem To Need Data Processing"

**Key Protocols by Layer:**
- **Application (7):** HTTP, HTTPS, FTP, SMTP, DNS, DHCP, Telnet, SSH
- **Transport (4):** TCP, UDP
- **Network (3):** IP, ICMP, IGMP, routing protocols (RIP, OSPF, BGP)

#### 3. TCP vs UDP - Know This Cold

**TCP - Transmission Control Protocol:**
- Connection-oriented (three-way handshake)
- Reliable delivery with acknowledgments
- Ordered delivery (sequence numbers)
- Flow control and congestion control
- Slower but guaranteed delivery
- **Uses:** HTTP/HTTPS, FTP, SMTP, SSH, Telnet

**UDP - User Datagram Protocol:**
- Connectionless (no handshake)
- No reliability guarantees
- No ordering guarantees
- Faster, less overhead
- **Uses:** DNS, DHCP, VoIP, video streaming, online gaming, SNMP

**When to Use Each:**
- TCP: When every bit matters (files, web pages, emails)
- UDP: When speed matters more than perfection (live video, gaming)

#### 5. TCP Three-Way Handshake

**Must be able to draw and explain:**

```
Client                          Server
  |                                |
  |----------SYN (seq=x)---------->|  Step 1: Client initiates
  |                                |
  |<------SYN-ACK (seq=y,ack=x+1)--|  Step 2: Server responds
  |                                |
  |----------ACK (ack=y+1)-------->|  Step 3: Client confirms
  |                                |
  |         Connection Established |
```

**Connection Termination (Four-way handshake):**
- FIN, ACK, FIN, ACK sequence

#### 6. Network Troubleshooting - Systematic Approach

**Always Start from Layer 1 and Work Up:**

**Layer 1 - Physical:**
- Check cables, connections, link lights
- Test with different cable/port
- Verify hardware is powered on

**Layer 2 - Data Link:**
- Check MAC address
- Verify switch port is enabled
- Check VLAN configuration
- Use `arp -a` command

**Layer 3 - Network:**
- Verify IP configuration (`ipconfig /all`)
- Check routing table (`route print` or `netstat -r`)
- Ping default gateway
- Check for IP conflicts

**Layer 4+ - Transport/Application:**
- Test specific services (ping, telnet to port)
- Check firewall rules
- Verify application configuration
- Check DNS (`nslookup`)
